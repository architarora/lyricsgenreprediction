# Audio Features Models

```python
import pandas as pd
import numpy as np
```

## Below are the steps we used to merge the two datasets to get all of the features/genres into one datset.

```python
poprock = pd.read_csv('additional_features_pop_rock_miheer.csv')
poprock.T
```

```python
pop = poprock.loc[poprock['genre'] == "Pop"]
rock = poprock.loc[poprock['genre'] == "Rock"].sample(325)
poprock = pd.concat([pop, rock])
poprock['genre'].value_counts()
```

```python
metalhip = pd.read_csv('additional_features_metal_hip_archit.csv')
metalhip = metalhip.rename(index = str, columns = {"Unnamed: 0" : "song_name", "Unnamed: 0.1" : "index"})
metalhip.T
```

```python
metal = metalhip.loc[metalhip['genre'] == "Metal"].sample(325)
hip = metalhip.loc[metalhip['genre'] == 'Hip-Hop'].sample(325)
metalhip = pd.concat([metal, hip])
metalhip['genre'].value_counts()
```

```python
genres = pd.concat([poprock, metalhip])
genres['genre'].value_counts()
```

```python
genres.T
```

### Removing unnecessary columns which have no effect on the type of genre. Found these by doing some feature engineering and from searching up online which metrics have the most effect on the type of the genre the music falls under.

```python
genresFiltered = genres.drop(columns=['index','analysis_url', 'id', 'key', 'mode', 'track_href', 'type', 'uri', 'lyrics', 'artist', 'song_artist', 'time_signature', "instrumentalness", "duration_ms", 'valence'])
genresFiltered.T
```

```python
genreSeries = genresFiltered['genre']
features = genresFiltered.drop(columns = ['song_name', 'genre'])
features.T
```

### Creating the design matrix for our song metrics.

```python
featuresList = features.values
featuresList
```

### Creating our label matrix by encoding each genre to a number, which is the format scikit learn needs it in.

```python
genreTOindex = {"Pop" : 0, "Rock" : 1, "Metal" : 2, "Hip-Hop" : 3}
indexTOgenre = {0 : "Pop", 1 : "Rock", 2 : "Metal", 3 : "Hip-Hop"}
target = genreSeries.map(lambda x: genreTOindex[x])
target
```

## Naive Bayes Training and Prediction
Using Cross Validation to test the performance of our model.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(featuresList, target, test_size=0.33)

from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)

expected = y_test
predicted = model.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.accuracy_score(expected, predicted))
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

### After some feature engineering, our best accuracy for Gaussian Naive Bayes around 0.6307. Considering we had very few training examples after our data cleaning and getting information through Spotify, we were pretty impressed with this number.

Once we added a max_depth parameter to prevent overiftting, the mean cross validation score increases to a number very close to the Gaussian Naive Bayes model earlier.

```python
clf = RandomForestClassifier(n_estimators=10, max_depth = 4)
```

```python
scores = cross_val_score(clf, featuresList,target, cv=10)
```

```python
scores.mean()
```

This gives us a pretty high accuracy as opposed to the other methods used above.

Testing if SVM with linear kernel would affect performance.

```python
clf = svm.SVC(kernel='linear', C=1, gamma='scale').fit(X_train, y_train)
```

```python
scores = cross_val_score(clf, featuresList,target, cv=10)
```

```python
scores.mean()
```

### Accuracy is very similar to what lyrics predicted and what Naive Bayes was able to predict. In order to improve performance, the best course of action would be to get more training examples because after using several models and trying some feature engineering, the accuracy is not increasing by a significant margin.

## Neural Networks

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
```

```python
scaler.fit(featuresList)  
featuresList = scaler.transform(featuresList)  
# target = scaler.transform(target)
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=5)
```

```python
scores = cross_val_score(clf, featuresList,target, cv=10)
```

```python
scores.mean()
```

With a MLP (multi-layer perceptron) neural network, with hidden layer sizes ranging from 2-5, we were able to get roughly the same accuracy as with the other models. MLP Neural Networks don't account for possible correlations between features, which in this case makes sense since the features are not correlated.

### Conclusion about features: With many different types of training models and feature engineering, no model had a significant difference in accuracy vs. other models. The reason for accuracy being lower than we expected is the limitations of our training examples. With more training examples, we should definitely be able to boost accuracy. In additoin, not all of the features are independent, which these models all assume.

## Ensemble Methods: Using multiple classifiers to boost performance

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from itertools import product
from sklearn.ensemble import VotingClassifier
```

```python
clf1 = SVC(gamma='scale', kernel='linear', probability=True)
```

```python
clf2 = DecisionTreeClassifier(random_state=42, max_depth = 5)
```

```python
clf_nn = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=5)
```

```python
clf3 = KNeighborsClassifier(n_neighbors=7)
```

```python
clf4 = RandomForestClassifier(n_estimators=10, max_depth = 4)
```

```python
eclf = VotingClassifier(estimators=[('svc', clf1), ('dt', clf_nn),('rf', clf4)],voting='soft', weights=[2,1,2])
```

```python
eclf.fit(featuresList, target)
```

```python
scores = cross_val_score(eclf, featuresList,target, cv=10)
```

```python
scores.mean()
```

After using a series of ensemble methods with different classifers, we get the accuracy to be 0.6577 which is higher than what each of the three models predicted earlier.
